```python
import pandas as pd
url = ('https://raw.githubusercontent.com/rgrAndrade/AlquimiaDosImoveis/refs/heads/main/Dados/dados_hospedagem.json')
dados_hospedagem = pd.read_json(url)
dados_hospedagem.head()
```

```python
dados_hospedagem = pd.json_normalize(dados_hospedagem['info_moveis'])
dados_hospedagem
```

```python
colunas = list(dados_hospedagem.columns)
colunas
```

```python
dados_hospedagem = dados_hospedagem.explode(colunas[3:])
dados_hospedagem
```

```python
dados_hospedagem.reset_index(drop=True, inplace=True)
dados_hospedagem.head(3)
```

```python
dados_hospedagem.info()
```

```python
import numpy as np
```

```python
dados_hospedagem['max_hospedes'] = dados_hospedagem['max_hospedes'].astype(np.int64)
dados_hospedagem.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados_hospedagem[col_numericas] = dados_hospedagem[col_numericas].astype(np.int64)
dados_hospedagem.info()
```

```python
dados_hospedagem['avaliacao_geral'] = dados_hospedagem['avaliacao_geral'].astype(float)
dados_hospedagem.info()
```

```python
dados_hospedagem['preco']
```

```python
dados_hospedagem['preco'] = dados_hospedagem['preco'].astype(str) # Convert the 'preco' column to string type
dados_hospedagem['preco'] = dados_hospedagem['preco'].apply(lambda x: x.replace('', '').replace(',', '').strip()) # Now apply your lambda function for cleaning
```

```python
dados_hospedagem['preco'] = dados_hospedagem['preco'].astype(np.float64)
dados_hospedagem.info()
```

```python
dados_hospedagem[['taxa_deposito', 'taxa_limpeza']] = dados_hospedagem[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados_hospedagem[['taxa_deposito', 'taxa_limpeza']] = dados_hospedagem[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
dados_hospedagem.info()
```
